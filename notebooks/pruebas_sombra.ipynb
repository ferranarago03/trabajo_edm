```python
import pvlib
from pvlib.location import Location
import pandas as pd

# Valencia
latitude = 39.4699
longitude = -0.3763
tz = 'Europe/Madrid'

# Fecha y hora (puede ser una serie horaria)
times = pd.date_range('2025-06-12 08:00', '2025-06-12 20:00', freq='H', tz=tz)

# Crear objeto ubicación
valencia = Location(latitude, longitude, tz)

# Obtener posición solar
solar_position = valencia.get_solarposition(times)

print(solar_position[['apparent_elevation', 'azimuth']])
```
